```python
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression

df = pd.read_csv('Credit.csv')
df = df.drop(['Gender', 'Student', 'Married', 'Ethnicity'], axis = 1)
df
```

```python
df.plot(kind = 'scatter', x = 'Income', y = 'Limit', title = "Income against Limit")
```

```python
plot=sns.lmplot(y = 'Limit', x = 'Income', data = df)
```

```python
res=sns.residplot('Limit', 'Income', data = df)
res.set(xlabel = 'Limit', ylabel = 'Residuals')
```

```python
print("Income")
linReg = LinearRegression().fit(df[['Income']], df[['Limit']])
print('Intercept', linReg.intercept_, '; Slope:', linReg.coef_)

print("\n From analysis, predictor maybe suitable for SLR")
```
