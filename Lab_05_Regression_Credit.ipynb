```python
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression

df = pd.read_csv('Credit.csv')
df = df.drop(['Gender', 'Student', 'Married', 'Ethnicity'], axis = 1)
df
```

```python
df.plot(kind = 'scatter', x = 'Limit', y = 'Income', title = "Income against Limit")
```

```python
plot=sns.lmplot(y = 'Income', x = 'Limit', data = df)
```

```python
res=sns.residplot('Limit', 'Income', data = df)
res.set(xlabel = 'Limit', ylabel = 'Residuals')
```

```python
print("Income")
linReg = LinearRegression().fit(df.drop('Income', axis = 1),df.iloc[:, 1])
linReg
print('Intercept', linReg.intercept_, '; Slope:', linReg.coef_)
```
